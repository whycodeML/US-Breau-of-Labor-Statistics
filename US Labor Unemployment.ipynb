```python
import requests
import json
from prettytable import PrettyTable
import pandas as pd
```

```python
headers = {'Content-type': 'application/json'}
data_us = json.dumps({"seriesid": ['LNS14000000'],"startyear":"2012", "endyear":"2021"})
p_us = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data_us, headers=headers)
json_data_us = json.loads(p_us.text)

for series in json_data_us['Results']['series']:
    x=prettytable.PrettyTable(["series id","year","period","periodName","value","footnotes"])
    seriesId = series['seriesID']
    for item in series['data']:
        year = item['year']
        period = item['period']
        periodName = item['periodName']
        value = item['value']
        footnotes=""
        for footnote in item['footnotes']:
            if footnote:
                footnotes = footnotes + footnote['text'] + ','
        if 'M01' <= period <= 'M12':
            x.add_row([seriesId,year,period,periodName,value,footnotes[0:-1]])
    output = open(seriesId + '.csv','w')
    output.write (x.get_string())
    output.close()
    
    #Storing JSON output in a file which would be downloaded in the desktop
```

```python
json_data_us
```

```python
#Loading Data in Dataframe

df_us = pd.json_normalize(json_data_us['Results']['series'],record_path=['data'])
df_us
```

```python
df_us['value']=df_us['value'].astype(float)
df_us.dtypes
```

```python
df_us_mean = round(df_us.groupby(by='year',as_index=False).mean(),1)
df_us_mean = df_us_mean.sort_values(by='year', ascending= False)
df_us_mean
```

```python
#Exporting monthly and yearly data in two separte sheets

with pd.ExcelWriter('US Unemployment Rate.xlsx') as writer:
    df_us.to_excel(writer, sheet_name='Monthly - Unemployment Rate',index=False)
    df_us_mean.to_excel(writer, sheet_name='Yearly - Unemployment Rate', index= False)
```
